```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.vinology.com/wine-terms/')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
wine_voc = []
```

```python
soup.find_all('strong')[0].contents
```

```python
import bs4

wine_voc = []

tags = soup.find_all('strong')
for tag in tags:
    word = tag.contents[0]
    if type(word) == bs4.element.NavigableString:
        word = word.strip()
        if word != '—':
            word = word.rstrip('\xa0—')
            word = word.rstrip(' |')
            word = word.lower()
            wine_voc.append(word)
```

```python
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

relevant_tokens = []
for i, data in dataset.iterrows():
    filtered = []
    description_tokens = word_tokenize(data['description'])
    filtered = [w for w in description_tokens if w not in stop_words]
    relevant_tokens.append(filtered)
    
dataset['filtered_description'] = relevant_tokens
```
